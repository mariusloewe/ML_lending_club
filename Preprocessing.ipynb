#### Removing NaN

```python
# remove columns with more than 95% of missing values

df = df.loc[:, df.isna().mean() < .95]
```

```python
# remove rows with more than 100 missing values
df = df.dropna(axis=0, thresh=100)
```

```python
#see here that almost all rows have at least one missing value, some of which are categorical and difficult ti impute
df.loc[:, df.isna().any()]
```

```python
#Decide to drop all columns with missing values - have 82 columns left
df = df.dropna(axis=1)
```

```python
df.dtypes
```

```python
# categorising grade (G is lowest and A is highest)
# categorising subgrade (1 is low and 5 is high)


df['grade'] = df['grade'].map({'G':'1', 'F':'2', 'E': '3', 'D':'4', 'C': '5', 'B': '6', 'A':'7'})
df['sub_grade'] = df['sub_grade'].map({'A1':'31', 'A2':'32', 'A3':'33', 'A4':'34', 'A5':'35', 'B1':'26', 'B2':'27', 'B3':'28', 'B4':'29', 'B5':'30', 'C1':'21',
       'C2':'22', 'C3':'23', 'C4':'24', 'C5':'25', 'D1':'16', 'D2':'17', 'D3':'18', 'D4':'19', 'D5':'20', 'E1':'11', 'E2':'12',
       'E3':'13', 'E4':'15', 'E5':'15', 'F1':'6', 'F2':'7', 'F3':'8', 'F4':'9', 'F5':'10', 'G1':'1', 'G2':'2', 'G3':'3',
       'G4':'4', 'G5':'5'})
```

## we can probably remove the 'grade' column since subgrade holds the same information & more

```python
df
```

```python
#convert issue_d to datetime
df['issue_d'] = pd.to_datetime(df.issue_d)
```

```python
# turn boolean columns to binary
df['debt_settlement_flag'] = df['debt_settlement_flag'].map({'N':'0', 'Y':'1'})

df['initial_list_status'] = df['initial_list_status'].map({'f':'0', 'w':'1'})

df['hardship_flag'] = df['hardship_flag'].map({'N':'0', 'Y':'1'})

df['pymnt_plan'] = df['pymnt_plan'].map({'n':'0', 'y':'1'})
```

```python
# One Hot Encode categorical values

home_ownership = pd.get_dummies(df['home_ownership'], prefix='home_ownership_', drop_first=False)
df = pd.concat([df, home_ownership], axis=1)
        
verification_status = pd.get_dummies(df['verification_status'], prefix='verification_status_', drop_first=False)
df = pd.concat([df, verification_status], axis=1)

application_type = pd.get_dummies(df['application_type'], prefix='application_type_', drop_first=False)
df = pd.concat([df, application_type], axis=1)

disbursement_method = pd.get_dummies(df['disbursement_method'], prefix='disbursement_method_', drop_first=False)
df = pd.concat([df, disbursement_method], axis=1)

loan_status = pd.get_dummies(df['loan_status'], prefix='loan_status_', drop_first=False)
df = pd.concat([df, loan_status], axis=1)

purpose = pd.get_dummies(df['purpose'], prefix='purpose_', drop_first=False)
df = pd.concat([df, purpose], axis=1)

term = pd.get_dummies(df['term'], prefix='term_', drop_first=False)
df = pd.concat([df, term], axis=1)

addr_state = pd.get_dummies(df['addr_state'], prefix='addr_state_', drop_first=False)
df = pd.concat([df, addr_state], axis=1)
```

```python
#drop categorical columns - are already to one hot encoded
df = df.drop(['home_ownership','verification_status','application_type','disbursement_method', 'loan_status',
             'purpose', 'term', 'addr_state'], axis=1)
```
